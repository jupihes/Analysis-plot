```python
import datetime as dt
def ymd_convert(x):
    '''
    Convert dates in the imported UFO data.
    Clean entries will look like YYYMMDD. If they're not clean, return NA.
    '''
    try:
        cnv_dt = dt.datetime.strptime(str(x), '%Y-%M-%D')
    except ValueError:
        cnv_dt = np.nan
        
    return cnv_dt

d = dt.datetime.strptime(df.SDATE, '%Y-%m-%d')
df['Time']= dt.date.strftime(d, "%m/%d/%y")
#df['Time'] = df.SDATE.map(ymd_convert)
df.head(4)
```
